1) Grafique las curvas de las series de Taylor de varios ordenes para los siguientes casos 


a. $f(x)=ln(x)$

Minimo de 4to o 5to orden

xo=1


$   f(x)=ln(x) \\
    f(x)'=\frac{1}{x}\\
    f(x)''=-x^{-2}\\
    f(x)'''=2x^{-3}\\
    f(x)''''=-6x^{-4}\\
    f(x)'''''=24x^{-5}\\
    P(x)=ln(1)+(x-1)-\frac{1}{2!}(x-1)^2+\frac{1}{3!}(x-1)^3-\frac{6}{4!}(x-1)^4+\frac{24}{5!}(x-1)^5$

```python
import numpy as np 
import matplotlib.pyplot as plt
import math 

x = np.linspace(0.1, 1.7, 200)

f = np.log(x)

x1 = (x - 1)
x2 = x1 - (x - 1)**2 / 2
x3 = x1 - (x - 1)**2 / 2 + (x - 1)**3 / 3
x4 = x1 - (x - 1)**2 / 2 + (x - 1)**3 / 3 - (x - 1)**4 / 4
x5 = x1 - (x - 1)**2 / 2 + (x - 1)**3 / 3 - (x - 1)**4 / 4 + (x - 1)**5 / 5

plt.figure(figsize=(8,6))
plt.plot(x, f, label='y = ln(x)', color='#FF00FF', linewidth=2)
plt.plot(x, x1, label='1er orden', color='red', linestyle='--')
plt.plot(x, x2, label='2 orden', color='orange', linestyle='--')
plt.plot(x, x3, label='3 orden', color='green', linestyle='--')
plt.plot(x, x4, label='4 orden', color='blue', linestyle='--')
plt.plot(x, x5, label='5 orden', color='purple', linestyle='--')

plt.axvline(0, color='black', linewidth=0.9)
plt.axhline(0, color='black', linewidth=0.9)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

con el punto xo=10

```python
import numpy as np 
import matplotlib.pyplot as plt
import math 

x = np.linspace(-1,20,200)

f = np.log(10) 

x1 = np.log(10) + (1/10)*(x - 10)
x2 = x1 - (1/(2*10**2))*(x - 10)**2
x3 = x2 + (1/(3*10**3))*(x - 10)**3
x4 = x3 - (1/(4*10**4))*(x - 10)**4
x5 = x4 + (1/(5*10**5))*(x - 10)**5


plt.figure(figsize=(8,6))
plt.plot(x, f * np.ones_like(x), label='y = ln(10)', color='#FF00FF', linewidth=2)
plt.plot(x, x1, label='1er orden', color='red', linestyle='--')
plt.plot(x, x2, label='2 orden', color='orange', linestyle='--')
plt.plot(x, x3, label='3 orden', color='green', linestyle='--')
plt.plot(x, x4, label='4 orden', color='blue', linestyle='--')
plt.plot(x, x5, label='5 orden', color='purple', linestyle='--')
plt.scatter(10, 2.3, color='red', s=40, label='Punto (10,2.3)')

plt.axvline(0, color='black', linewidth=0.9)
plt.axhline(0, color='black', linewidth=0.9)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

b) $f(x)=cos(x) , x_o=0$

$   f(x)=cos(x)\\
    f(x)'=-sen(x)\\
    f(x)''=-cos(x)\\
    f(x)'''=sen(x)\\
    f(x)''''=cos(x)\\
    f(x)'''''=-sen(x)\\       
    P(x)=1-sen(1)(x-1)-\frac{cos(1)}{2}(x-1)^2+\frac{sen(1)}{6}(x-1)^3+\frac{cos(1)}{24}(x-1)^4-\frac{sen(1)}{120}(x-1)^5
$


```python
import numpy as np 
import matplotlib.pyplot as plt
import math 

x = np.linspace(-5,5,200)

f = np.cos(x)

x1 = 1-(np.sin(1)*(x-1))
x2 = x1 - (np.cos(1)/2)*(x - 1)**2
x3 = x2 + (np.sin(1)/6)*(x - 1)**3
x4 = x3 - (np.cos(1)/24)*(x - 1)**4
x5 = x4 + (np.sin(1)/(120))*(x - 1)**5


plt.figure(figsize=(12,6))
plt.plot(x, f, label='y=cos(x)', color='#FF00FF', linewidth=2)
plt.plot(x, x1, label='1er orden', color='red', linestyle='--')
plt.plot(x, x2, label='2 orden', color='orange', linestyle='--')
plt.plot(x, x3, label='3 orden', color='green', linestyle='--')
plt.plot(x, x4, label='4 orden', color='blue', linestyle='--')
plt.plot(x, x5, label='5 orden', color='purple', linestyle='--')

plt.axvline(0, color='black', linewidth=0.9)
plt.axhline(0, color='black', linewidth=0.9)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.legend()
plt.grid(True)
plt.show()
```

c) $\frac{1}{x-1}, xo=\frac{1}{2}$

$   f(x)'=\frac{1}{(1-x)^2}\\
    f(x)''=\frac{2}{(1-x)^3}\\
    f(x)'''=\frac{6}{(1-x)^4}\\
    f(x)''''=\frac{24}{(1-x)^5}\\
    f(x)'''''=\frac{120}{(1-x)^6}\\
    P(x)= 2+4(x-\frac{1}{2})+8(x-\frac{1}{2})^2+16(x-\frac{1}{2})^3)+32(x-\frac{1}{2})^4 +64(x-\frac{1}{2})^5 
$


```python
import numpy as np 
import matplotlib.pyplot as plt
import math 

x = np.linspace(-1, 2, 400)
f = 1/(1 - x)


x1 = 2+4*(x-1/2)
x2 = x1 + (8)*(x - 1/2)**2
x3 = x2 + (16)*(x - 1/2)**3
x4 = x3 + (32)*(x - 1/2)**4
x5 = x4 + (64)*(x - 1/2)**5


plt.figure(figsize=(8,6))
plt.plot(x, f, label='y=1/(1-x)', color='#FF00FF', linewidth=2)
plt.plot(x, x1, label='1er orden', color='red', linestyle='--')
plt.plot(x, x2, label='2 orden', color='orange', linestyle='--')
plt.plot(x, x3, label='3 orden', color='green', linestyle='--')
plt.plot(x, x4, label='4 orden', color='blue', linestyle='--')
plt.plot(x, x5, label='5 orden', color='purple', linestyle='--')

plt.axvline(0, color='black', linewidth=0.9)
plt.axhline(0, color='black', linewidth=0.9)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-1, 2)
plt.ylim(-10, 10)
plt.legend()
plt.grid(True)
plt.show()
```

Ejercicio 2

Encuentre el polinomio de Lagrange para los siguientes datos y grafique:

a) (0,0),(30,0.5),(60,3/2),(90,1)


$L_0(x)=\frac{(x-30)(x-60)(x-90)}{(0-30)(0-60)(0-90)}=-\frac{(x-30)(x-60)(x-90)}{162000}$

$L_1(x)=\frac{(x-0)(x-60)(x-90)}{(30-0)(30-60)(30-90)}=\frac{x(x-60)(x-90)}{54000}$

$L_2(x)=\frac{(x-0)(x-30)(x-90)}{(60-0)(60-30)(60-90)}=-\frac{x(x-30)(x-90)}{54000}$

$L_3(x)=\frac{(x-0)(x-30)(x-60)}{(90-0)(90-30)(90-60)}=\frac{x(x-30)(x-60)}{162000}$

$P(x)=0\left(-\frac{(x-30)(x-60)(x-90)}{162000}\right)
+\frac{1}{2}\left(\frac{x(x-60)(x-90)}{54000}\right)
+3\sqrt{2}\left(-\frac{x(x-30)(x-90)}{54000}\right)
+1\left(\frac{x(x-30)(x-60)}{162000}\right)$

$P(x) = -6.054\times10^{-7}x^3 - 1.9943\times10^{-5}x^2 + 0.0178098x$



```python
import numpy as np 
import matplotlib.pyplot as plt
import math 

x = np.linspace(0, 100, 400)

f=-6.054e-7 * x**3 - 1.9943e-5 * x**2 + 0.0178098 * x

xp = [0, 30, 60, 90]
yp = [0, 0.5, math.sqrt(3)/2, 1]

plt.figure(figsize=(8,6))
plt.plot(x, f, label='Polinomio de Lagrange', color='#FF00FF', linewidth=2)
plt.scatter(xp, yp, color='red', label='Puntos dados', zorder=5)

plt.axvline(0, color='black', linewidth=0.9)
plt.axhline(0, color='black', linewidth=0.9)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0, 100)
plt.ylim(0, 1.2)
plt.legend()
plt.grid(True)
plt.title('Interpolacin de Lagrange')
plt.show()
```
